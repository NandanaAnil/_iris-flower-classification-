```python
iris['species'].value_counts()
```

```python
iris.replace({'species':{'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}}, inplace = True)
```

```python
iris.head(60)
```

```python
fig, ax = plt.subplots(figsize=(20,20))
sns.boxplot(data=iris, ax=ax)
```

```python
Q1 = iris.quantile(0.25)
Q3 = iris.quantile(0.75)

#Calculate IQR - Inter Quartile Range 
IQR = Q3 - Q1

#Determine the outlier boundaries 
lower_bound = Q1 - 1.5 * IQR 
upper_bound = Q3 + 1.5 * IQR

#Create a mask for data within the IQR boundaries 
mask = ~((iris < lower_bound) | (iris > upper_bound)).any(axis=1)

#Remove outliers 
cleaned_data = iris[mask]
```

```python
fig, ax = plt.subplots(figsize=(20,20))
sns.boxplot(cleaned_data, ax=ax)
```

```python
x = cleaned_data.drop('species', axis=1)
x
```

```python
y = cleaned_data['species']
y 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
train_data_predict = model.predict(x_train)
train_data_accuracy = accuracy_score(y_train, train_data_predict)
print("Accuracy of training data:", train_data_accuracy)
```

```python
test_data_predict = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test, test_data_predict)
print("Accuracy of testing data:", test_data_accuracy)
```

```python
cleaned_data.head()
```

```python
cm = confusion_matrix(test_data_predict, y_test)
cm
```

```python
sns.heatmap(cm, annot = True, annot_kws = {'size':20}, fmt = '.2f', cmap = 'Pastel1', square = True, 
           xticklabels = ['predicted_iris_setosa', 'predicted_iris_versicolor', 'predicted_iris_virginica'],
           yticklabels = ['actual_iris_setosa', 'actual_iris_versicolor', 'actual_iris_virginica'])
plt.xlabel('predicted_area')
plt.ylabel('actual_area')
plt.title('confusion_matrix')
plt.show()
```

```python
print(classification_report(test_data_predict, y_test))
```
